# Criar novas variáveis

var continua -> var categorica 




var categorica -> continua

```
dicionario_de_salarios = {
    'Menos de R$ 1.000/mês': 1000,
    'de R$ 1.001/mês a R$ 2.000/mês': 1500, 
    'de R$ 2.001/mês a R$ 3000/mês': 2500,
    'de R$ 3.001/mês a R$ 4.000/mês': 3500,
    'de R$ 4.001/mês a R$ 6.000/mês': 5000, 
    'de R$ 6.001/mês a R$ 8.000/mês': 7000,
    'de R$ 8.001/mês a R$ 12.000/mês': 10000,
    'de R$ 12.001/mês a R$ 16.000/mês': 14000,
    'de R$ 16.001/mês a R$ 20.000/mês': 18000,
    'de R$ 20.001/mês a R$ 25.000/mês': 22500, 
    'Acima de R$ 25.001/mês': 25000}
```

```
dados['salario']=dados["('P16', 'salary_range')"].map(dicionario_de_salarios)
```

```
#Fazendo a contagem por faixa salarial
dados.salario.value_counts().sort_index()
```

```
dados.salario.value_counts().sort_index().plot(kind='barh');
```

```
def idade(x):
        if x <= 24:
            return '[18, 24]'
        elif x <= 30:
            return '[25, 30]'
        elif x <= 40:
            return '[31, 40]'
        else:
            return  '[41, 50]'                
```

```
n=dados.shape[0]
```

```
dados['nova idade']=dados["('P1', 'age')"].apply(idade)
```

```
#Verificando a contagem da coluna criada
dados['nova idade'].value_counts()/n
```
